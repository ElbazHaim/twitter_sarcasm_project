```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd 
df = pd.read_csv("results.csv")

x = np.array([1, 2, 3])

series1 = df["SVM"]
series2 = df["RandomForest"]
series3 = df["LogisticRegression"]

bar_width = 0.25
fig, ax = plt.subplots()
ax.bar(x, series1, width=bar_width, label='SVM')
ax.bar(x + bar_width, series2, width=bar_width, label='Random Forest')
ax.bar(x + 2 * bar_width, series3, width=bar_width, label='Logistic Regression')

ax.set_xticks(x + bar_width)
ax.set_xticklabels(["Level 1", "Level 2", "Level 3"])
ax.set_xlabel('Process Level')

ax.set_ylabel('Accuracy')
ax.set_ylim(0.7, 1)
ax.set_title('Accuracy Comparison for Top Models')
ax.legend()
plt.show()
```

It looks like SVM is getting better as we get higher in process levels, while logistic regression is getting worse results as we get higher in process levels

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd 
df = pd.read_csv("percision_0_table.csv")

x = np.array([1, 2, 3])

series1 = df["SVM"]
series2 = df["RandomForest"]
series3 = df["LogisticRegression"]

bar_width = 0.25
fig, ax = plt.subplots()
ax.bar(x, series1, width=bar_width, label='SVM')
ax.bar(x + bar_width, series2, width=bar_width, label='Random Forest')
ax.bar(x + 2 * bar_width, series3, width=bar_width, label='Logistic Regression')

ax.set_xticks(x + bar_width)
ax.set_xticklabels(["Level 1", "Level 2", "Level 3"])
ax.set_xlabel('Process Level')

ax.set_ylabel('Nonsarcasm Percision')
ax.set_ylim(0.5, 1)
ax.set_title('Nonsarcasm Percision Comparison for Top Models')
ax.legend()
plt.show()
```
